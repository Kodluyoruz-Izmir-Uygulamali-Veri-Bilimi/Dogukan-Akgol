# Soru 8

Age, Embarked ve Fare değişkenlerini mean/median/mode değerlerini kullanarak değişken tipine uygun olacak şekilde doldurunuz.

```python
df_titanic_8 = pd.read_csv("train_missing.csv")
display(df_titanic_8)
```

```python
df_titanic_8.Age.isnull().sum(), df_titanic_8.Embarked.isnull().sum(), df_titanic_8.Fare.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp_mean = SimpleImputer(strategy='mean')
```

```python
df_titanic_8[["Age","Fare"]]
```

```python
imp_mean.fit(df_titanic_8[["Age","Fare"]])
```

```python
display(df_titanic_8)
```

```python
imp_mean.transform(df_titanic_8[["Age","Fare"]])
```

```python
df_titanic_8[["Age","Fare"]] = imp_mean.transform(df_titanic_8[["Age","Fare"]])
```

```python
df_titanic_8.Age.isnull().sum(), df_titanic_8.Fare.isnull().sum() 
```

```python
imp_mode = SimpleImputer(strategy='most_frequent')
```

```python
imp_mode.fit(df_titanic_8[["Embarked"]])
```

```python
imp_mode.fit_transform(df_titanic_8[["Embarked"]])
```

```python
df_titanic_8[["Embarked"]] = imp_mode.fit_transform(df_titanic_8[["Embarked"]])
```

```python
df_titanic_8[["Age","Fare","Embarked"]].isnull().sum()
```
