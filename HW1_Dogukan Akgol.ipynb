```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

data = {"Brand": ["Huawei","HP","Lenovo","Huawei","Huawei"],
		"Model": ["Matebook D15","340S G7","Ideapad3","Matebook D13","Matebook D14"],
		"Processor": ["Ryzen5","Intel i3","Ryzen3","intel i5","Ryzen5"],
		"RAM(GB)": [8, 8, 4, 8, 8],
		"Price(TL)": [4999, 5339.99, 4307.16, 7553, 4999],
        "ScreenSize(inch)": [15.6, 14, 14, 13, 14],
        "Weight(kg)": [1.53, 1.47, 1.42, 1.3, 1.38]}

df1 = pd.DataFrame(data)
df1
```

```python
'''
Feature(Attribute) Types:
Brand---- Categorical | Nominal
Model---- Categorical | Nominal
Processor---- Categorical | Nominal
RAM(GB)---- Numeric | Ratio
Price(TL)---- Numeric | Ratio
ScreenSize(inch)---- Numeric | Ratio
Weight(kg)---- Numeric | Ratio
'''
```

```python
df1.describe(include='all')
```

```python
df2 = pd.DataFrame(data, columns=['RAM(GB)', 'Price(TL)', 'ScreenSize(inch)', 'Weight(kg)'])
green_diamond = dict(markerfacecolor='g', marker='D')
boxplot = df2.boxplot(column=['Price(TL)'], flierprops=green_diamond)
boxplot.set_title('Boxplot for Price(TL) Feature')
```

```python
#Quantiles------------------
Q =[]

Q1 = df1["Price(TL)"].quantile(0.25)
Q2=df1["Price(TL)"].quantile(0.50)
Q3 = df1["Price(TL)"].quantile(0.75)

IQR = Q3 - Q1
Q.append(Q1)
Q.append(Q2)
Q.append(Q3)
Q.append(IQR)
Q
```
